```python
import pandas as pd
import numpy as np
import datetime
import calendar

import tensorflow as tf

# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
Data_visit_1H = pd.read_pickle("../../data/Data_visit-20171226-5")

Data_sub_1H = pd.read_pickle("../../data/Data_sub-20171226-5")
```

```python
Data_sub_1H.columns.tolist()
```

```python
learning_end_date = datetime.date(2017,3, 13)
evaluation_end_date = datetime.date(2017,4, 22)

# Reservations data are horrible before that date, trying to improve by removing that period
relevant_start_date = datetime.date(2016,11, 1)


# Change this value to train with or without reservations features

col = [c for c in Data_visit_1H if c not in ['air_store_id_air_0ead98dd07e7a82a', 
                                             'air_store_id_air_229d7e508d9f1b5e', 
                                             'air_store_id_air_2703dcb33192b181', 
                                             'air_store_id_air_b2d8bc9c88b85f96', 
                                             'air_store_id_air_cb083b4789a8d3a2', 
                                             'air_store_id_air_cf22e368c1a71d53', 
                                             'air_store_id_air_d0a7bd3339c3d12a' ,
                                             'air_store_id_air_d63cfa6d6ab78446',
                                             'id', 
                                             'air_store_id',
                                             'visit_date',
                                             'visitors', 
                                             'DoW', 
                                             'air_area_name', 
                                             'city', 
                                             'prefecture']]

# Create the Test set

A = Data_sub_1H.copy()
X_test =  A[col].fillna(-1).as_matrix()
#Id_test = A[['air_store_id', 'visit_date']]

# Create the learning set

A = Data_visit_1H
#A = A[(Data_visit_1H['visit_date'] <  learning_end_date)]

X_learn = A[col].fillna(-1).as_matrix()
y_learn = A[['visitors']].as_matrix()
#Id_learn = A[['air_store_id', 'visit_date']]

# Create the evaluation set

A = Data_visit_1H
#A = A[(Data_visit_1H['visit_date'] >=  learning_end_date) & (Data_visit_1H['visit_date'] <=  evaluation_end_date)]
    
# Remove the lines with missing visits values
X_eval = A[col].fillna(-1).as_matrix()
y_eval = A[['visitors']].as_matrix()
#Id_learn =  A[['air_store_id', 'visit_date']] 
```

#### Build the neural network

```python
from tensorflow.contrib.layers import dropout
from tensorflow.contrib.layers import batch_norm
from tensorflow.contrib.layers import fully_connected

reset_graph()

n_features = X_learn.shape[1]

n_neurons_1 = 300
n_neurons_2 = 100
learning_rate = 0.015
dropout_rate = 0.5

X = tf.placeholder(tf.float32, shape=[None, n_features], name="X")
y = tf.placeholder(tf.float32, shape = [None,1], name = "y")
y_log1p = tf.log1p(y)

is_training = tf.placeholder_with_default(False, shape=[])

bn_params = {
    'is_training': is_training,
    'decay': 0.99,
    "updates_collections": None,
    "scale": True
}

with tf.name_scope("dnn"):
    hidden1 = fully_connected(X, n_neurons_1, 
                              normalizer_fn = batch_norm,
                              normalizer_params = bn_params,
                              activation_fn=tf.nn.elu, 
                              scope="hidden1")
    hidden1_drop = tf.layers.dropout(hidden1, dropout_rate, training=is_training)
    
    hidden2 = fully_connected(hidden1_drop, n_neurons_2, 
                              activation_fn=tf.nn.elu, 
                              normalizer_fn = batch_norm,
                              normalizer_params = bn_params,
                              scope="hidden2")
    hidden2_drop = tf.layers.dropout(hidden2, dropout_rate, training=is_training)

    
    output = fully_connected(hidden2_drop, 1,
                            normalizer_fn = batch_norm,
                            normalizer_params = bn_params,
                            activation_fn = None)

    
    
# Labels are inputed as logp1(y)
with tf.name_scope("loss"):
    msle_error = tf.losses.mean_squared_error(y_log1p,output)
    rmsle_error = tf.sqrt(msle_error)
       
with tf.name_scope("train"):
    optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate)
    training_op = optimizer.minimize(msle_error)  
    
pred = tf.expm1(output)
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

### Run the neural network

```python
n_epochs = 150
batch_size = 100

LOAD_FILE = "../../data/DNN-2-20171226-5d.ckpt"
SAVE_FILE = "../../data/DNN-2-20171226-5e.ckpt"

best_error = np.infty
checks_without_progress = 0
max_checks_without_progress = 35

with tf.Session() as sess:
    init.run()
    
    # To continue the training
    saver.restore(sess, LOAD_FILE)
    
    for epoch in range(n_epochs):
        rnd_idx = np.random.permutation(len(X_learn))
        for rnd_indices in np.array_split(rnd_idx, len(X_learn) // batch_size):
            X_batch, y_batch = X_learn[rnd_indices], y_learn[rnd_indices]
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, is_training: True})
        if epoch % 1 == 0:
            eval_error = rmsle_error.eval(feed_dict={X: X_eval, y: y_eval, is_training: False})
            print(epoch, "Eval accuracy:", eval_error)
            
            if eval_error < best_error:
                    #Saving the last version of the network
                    checks_without_progress = 0
                    save_path = saver.save(sess, SAVE_FILE)
                    best_error = eval_error

            else: 
                checks_without_progress += 1
                if checks_without_progress > max_checks_without_progress:
                    print("Early stopping!")
                    stopping = True
                    break

with tf.Session() as sess:
    saver.restore(sess, SAVE_FILE)
    print("Final eval error: {:.3f}".format(best_error))   
```
