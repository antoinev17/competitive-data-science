```python
import pandas as pd
import inspect
import os
dirname = os.path.dirname(os.path.abspath(inspect.stack()[0][1]))

TRAIN_DATA_PATH = "data/train.csv"
TEST_DATA_PATH = "data/test.csv"


def load_rain_data(data_path=TRAIN_DATA_PATH):
    data_path = os.path.join(dirname, TRAIN_DATA_PATH)
    return pd.read_csv(data_path)
```

### Remove all data with expected bigger than 1m

```python
import numpy as np

#rain_cleaned = rain[rain.Expected < np.percentile(rain.Expected,95)]
rain_cleaned = rain[rain.Expected < 1000]

%matplotlib inline 
import matplotlib.pyplot as plt
rain_cleaned.Expected.hist(bins=50,figsize=(20,150))
plt.show()
```

```python
rain_cleaned.describe()
```

### Create the test/training sets

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(rain_cleaned, test_size = 0.2, random_state=42)

rain_train = train_set.drop("Expected", axis = 1)
rain_train_labels = train_set["Expected"].copy()

rain_test = test_set.drop("Expected", axis = 1)
rain_test_labels = test_set["Expected"].copy()
```

### Transformation pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer

num_pipeline = Pipeline([
    ('imputer', Imputer(strategy="median")),
    ('std_scaler', StandardScaler())   
])

rain_transform = num_pipeline.fit_transform(rain_train)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(rain_transform, rain_train_labels)
```

```python
from sklearn.metrics import mean_squared_error
rain_predictions = lin_reg.predict(rain_transform)
lin_mse = mean_squared_error(rain_train_labels, rain_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

### Random forest

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(rain_transform, rain_train_labels) 
```

```python
rain_predictions = tree_reg.predict(rain_transform)
tree_mse = mean_squared_error(rain_train_labels, rain_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
rain_predictions = tree_reg.predict(num_pipeline.transform(rain_test))
tree_mse = mean_squared_error(rain_test_labels, rain_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```
